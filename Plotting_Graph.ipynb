```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
def graph_plot(df,flag,title):
    clf = ['a','aa','rr','a','aa','rr']
    labels = df.iloc[:,0]
    if flag==1:
        data1 = df.iloc[:,1]
        data2 = df.iloc[:,3]
        data3 = df.iloc[:,5]
        data4 = df.iloc[:,7]
        data5 = df.iloc[:,9]
        data6 = df.iloc[:,11]
    elif flag==2:
        data1 = df.iloc[:,2]
        data2 = df.iloc[:,4]
        data3 = df.iloc[:,6]
        data4 = df.iloc[:,8]
        data5 = df.iloc[:,10]
        data6 = df.iloc[:,12]
    elif flag==3 or flag==5:
        data1 = df.iloc[:,1]
        data2 = df.iloc[:,3]
        data3 = df.iloc[:,5]
        data4 = df.iloc[:,7]
        data5 = df.iloc[:,9]
    elif flag==4 or flag==6:
        data1 = df.iloc[:,2]
        data2 = df.iloc[:,4]
        data3 = df.iloc[:,6]
        data4 = df.iloc[:,8]
        data5 = df.iloc[:,10]
   
         
    if flag==1 or flag == 2:
        groups = [data1, data2,data3,data4, data5,data6]
    elif flag==3 or flag == 4 or flag == 5 or flag == 6:
         groups = [data1, data2,data3,data4, data5]
    margin = 0.009
    width = (1.-2.*margin)/10
    ind = np.arange(9)
    s = plt.subplot(1,1,1)
    plt.legend(labels)
    for num, vals in enumerate(groups):
        # The position of the xdata must be calculated for each of the two data series
        xdata = ind+margin+(num*width)
        # Removing the "align=center" feature will left align graphs, which is what
        # this method of calculating positions assumes
        gene_rects = plt.bar(xdata, vals, width)

        
        
    if (flag==1 or flag==2 ):
        plt.legend(['KNN','SVM','DTC','ABC','LR','GNB'])
        
    elif (flag==3 or flag==4 or flag ==5 or flag == 6):
        plt.legend(['KNN','SVM','DTC','ABC','LR'])
      
   
   
    
    
    # You should no longer need to manually set the plot limit since everything 
    # is scaled to one.
    # Also the ticks should be much simpler now that each group of bars extends from
    # 0.0 to 1.0, 1.0 to 2.0, and so forth and, thus, are centered at 0.5, 1.5, etc.
    s.set_xticks(ind+0.5)
    s.set_xticklabels(labels,rotation=90)
    s.set_title(title)
```

```python
df = pd.read_csv('Plotting_Graph/ML_Project - Plot_Classification_default.csv',skiprows=1)
graph_plot(df,1,"Classification Training Accuracy (default)")
```

```python
graph_plot(df,2,"Classification Testing Accuracy (default)")
```
